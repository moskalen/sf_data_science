### Задание 5.1 (1 балл)

Давайте сделаем предобработку данных. 

Для начала расширим датасет с помощью методов Feature Engineering. Создайте следующие признаки:
* `BalanceSalaryRatio` — отношение баланса на счетах к заработной плате клиента;
* `TenureByAge` — отношение времени пользования услугами банка к возрасту клиента;
* `CreditScoreGivenAge` — отношение кредитного рейтинга к возрасту клиента.

Затем перекодируйте оставшиеся текстовые категориальные признаки в числовые бинарные столбцы.

Выведите основные статистические характеристики полученных признаков.

**Критерии оценивания:**
- Задание выполнено верно, учтены все условия (**1 балл**): 
    * приведён код для создания указанных в задании признаков;
    * приведён код для кодирования категориальных признаков в числовые бинарные столбцы;
    * приведён код для расчёта основных статистических характеристик полученных признаков.

- Задание выполнено неверно, не учтено одно или несколько условий (**0 баллов**).

````python
# Добавим признак `BalanceSalaryRatio` — отношение баланса на счетах к заработной плате клиента
if 'BalanceSalaryRatio' not in churn_data:
    churn_data['BalanceSalaryRatio'] = churn_data.apply(
        lambda row: row['Balance'] / row['EstimatedSalary'], axis=1
    )

# Добавим признак `TenureByAge` — отношение времени пользования услугами банка к возрасту клиента
if 'TenureByAge' not in churn_data:
    churn_data['TenureByAge'] = churn_data.apply(
        lambda row: row['Tenure'] / row['Age'], axis=1
    )

# Добавим признак `CreditScoreGivenAge` — отношение кредитного рейтинга к возрасту клиента
if 'CreditScoreGivenAge' not in churn_data:
    churn_data['CreditScoreGivenAge'] = churn_data.apply(
        lambda row: row['CreditScore'] / row['Age'], axis=1
    )

# преобразуем пол в бинарный признак: 1 - Male, 0 - Female
if churn_data.dtypes['Gender'] == 'object':
    churn_data['Gender'] = churn_data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
    
churn_data.describe()
````

Разделим исходный набор данных на матрицу наблюдений `X` (фичи) и столбец ответов `y` (таргет). 

````python
X = churn_data.drop("Exited", axis=1)
y = churn_data["Exited"]
````

Прежде чем перейти к этапу разделения выборки на тренировочную и тестовую, давайте посмотрим на сбалансированность данных. Для этого воспользуемся методом `value_counts()` и выведем количество наблюдений каждого из классов в долевом соотношении от общего количества наблюдений:

````python
y.value_counts(normalize=True)
````

Очевидно, что классы не сбалансированы. Ушедших пользователей 32 %, в то время как лояльных — 68 %.


**К чему это может привести?**

При разделении набора данных на тренировочный и тестовый может возникнуть такая ситуация, что в тренировочную выборку попадут только лояльные клиенты или их большая часть. Тогда модель может не научиться определять вероятность оттока и качество на тестовой выборке будет неудовлетворительным.

Или противоположная ситуация: в тестовую выборку попадут только лояльные клиенты. Тогда модель будет показывать отличное качество на тестовой выборке, но в реальных условиях она может оказаться бесполезной.

То есть нам нужно сделать так, чтобы в тренировочной и тестовой выборке было одинаковое соотношение лояльных и ушедших клиентов. Такое разбиение выборки называется **стратифицированным**. 

Функция `train_test_split()` из библиотеки `sklearn` умеет производить такое разбиение. Для этого в ней предусмотрен параметр `stratify`, который нужно выставить в значение признака, который должен иметь одинаковые соотношения в тренировочной и тестовой выборке. У нас этот признак - это целевой признак y. 

Выполним разбиение, установив значение параметра `random_state` на 0:


````python
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y, 
    stratify=y, 
    random_state=0
)
print('Train shape: {}'.format(X_train.shape))
print('Test shape: {}'.format(X_test.shape))
````

Проверим, что число лояльных и ушедших клиентов примерно одинаково в каждой из выборок:

````python
print('Train :')
display(y_train.value_counts(normalize=True))
print('\n')
print('Test :', )
display(y_test.value_counts(normalize=True))
````

Соотношения одинаковы, а значит, мы можем перейти к следующему шагу.

````python
#Создаем объект класса логистическая регрессия
log_reg = linear_model.LogisticRegression(
    solver='sag', #алгоритм оптимизации
    random_state=42,
    max_iter=1000,
    #penalty='l1' <-- не работает при L1-регуляризации
)

#Обучаем модель, минимизируя logloss
log_reg.fit(X_train_scaled, y_train)

#Делаем предсказание для тренировочной выборки
y_train_pred = log_reg.predict(X_train_scaled)

#Делаем предсказание для тестовой выборки
y_test_pred = log_reg.predict(X_test_scaled)

#Вывод отчет о метриках классификации
print('F1 score на тренировочной выборке: {:.2f}'.format(metrics.f1_score(y_train, y_train_pred)))
print('F1 score на тестовой выборке: {:.2f}'.format(metrics.f1_score(y_test, y_test_pred)))
````

Промежуточные выводы: 
* модель при $F_1$=0.47 на тестовых данных не удовлетворяет нашим требованиям и является недообученной.
* код не работает при $L_1$-регуляризации, поэтому оставляем `penalty='l1'` (параметр по-умолчанию)

Попробуем найти оптимальный параметр `C`, а также оптимальное пороговое значение для разделения классов

> **Примечание.** Учтите, что для предсказания необходимо преобразовать данные к тому формату, на котором обучалась модель, то есть:
* добавить признаки дополнительные признаки; 
* произвести кодирование категориальных признаков;
* произвести масштабирование признаков (если ваша модель обучалась на масштабированных данных) с помощью того же скейлера, который вы использовали при обучении модели;
* сгенерировать полиномиальные признаки.

**Критерии оценивания:**
- Задание выполнено верно, учтены все условия (**1 балл**): 
    * сформирован DataFrame на основе предоставленного словаря;
    * произведены преобразования данных, необходимые для работы модели;
    * предсказана вероятность оттока клиента (вероятность принадлежности к классу 1).

- Задание выполнено неверно, не учтено одно или несколько условий (**0 баллов**).

````python
def get_class(client_info: dict, threshold: float = 0.42) -> (int, float):
    """
    Предсказывание оттока клиента моделью случайного леса
    
    Args:
        info (dict): информация о клиенте
        threshold (float): оптимальный порог вероятности. По-умолчанию, 0.42
        
    Returns:
        (int, float): (предсказанные класс: 1 - клиент уйдет, 0 - не уйдет, вероятность ухода)
    """
    data = pd.DataFrame(client_info)
    
    int_columns = ['CreditScore', 'Age', 'Tenure', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']
    for col in int_columns:
        data[col] = data[col].astype(np.int64)

    # Добавим признак `BalanceSalaryRatio` — отношение баланса на счетах к заработной плате клиента
    if 'BalanceSalaryRatio' not in data:
        data['BalanceSalaryRatio'] = data.apply(
            lambda row: row['Balance'] / row['EstimatedSalary'], axis=1
        )

    # Добавим признак `TenureByAge` — отношение времени пользования услугами банка к возрасту клиента
    if 'TenureByAge' not in data:
        data['TenureByAge'] = data.apply(
            lambda row: row['Tenure'] / row['Age'], axis=1
        )

    # Добавим признак `CreditScoreGivenAge` — отношение кредитного рейтинга к возрасту клиента
    if 'CreditScoreGivenAge' not in data:
        data['CreditScoreGivenAge'] = data.apply(
            lambda row: row['CreditScore'] / row['Age'], axis=1
        )

    # преобразуем пол в бинарный признак: 1 - Male, 0 - Female
    if data.dtypes['Gender'] == 'object':
        data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
        
    y_info_pred = rf.predict_proba(data)[:, 1][0]
    class_pred = 1 if y_info_pred >= threshold else 0
    
    return (class_pred, y_info_pred)

(class_pred, y_info_pred) = get_class({
    'CreditScore': [601.0],
    'Gender': ['Male'],
    'Age': [42.0],
    'Tenure': [1.0],
    'Balance': [98495.72],
    'NumOfProducts': [1.0],
    'HasCrCard': [1.0],
    'IsActiveMember': [0.0],
    'EstimatedSalary': [40014.76]
})

print(f"Оценка вероятности ухода Василия: {y_info_pred:0.2f}")
print(f"Василий {'уйдет' if class_pred == 1 else 'не уйдет'}.")
    
````

* клиент Василий вероятно откажется от услуг банка
